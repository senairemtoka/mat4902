```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

file_path = r"/Users/senairemtoka/Downloads/ghg_emissions_with_gdp_population (1).xlsx"
ghg_df = pd.read_excel(file_path)

```
